```python
import pandas as pd

# Load dataset
df1 = pd.read_csv("final_merged_dataset.csv")

# Drop unnecessary columns
df1.drop(columns=['Unnamed: 0', 'No','experiment_id','Machining Process'], inplace=True, errors='ignore')

# Optional: sort by timestamp if exists
if 'timestamp' in df1.columns:
    df1 = df1.sort_values(by='timestamp')

# Display basic info
print(df1.info())
print(df1.head())
```

```python
# Re-identify continuous columns
continuous_cols = df1.select_dtypes(include=['float64', 'int64']).columns.tolist()

# Remove any columns that are actually categorical despite being numeric (like 'tool_condition' if encoded)
non_continuous = ['tool_condition', 'material', 'machining_finalized', 'passed_visual_inspection']  # add if any
continuous_cols = [col for col in continuous_cols if col not in non_continuous]

# Verify
print(f"Total continuous features: {len(continuous_cols)}")
print(continuous_cols)

```

```python
from scipy.stats import skew, kurtosis

print(f"{'Column':35} | {'Skewness':>8} | {'Kurtosis':>9}")
print("-" * 60)

for col in continuous_cols:
    col_data = df1[col].dropna()
    sk = skew(col_data)
    ku = kurtosis(col_data, fisher=False)  # fisher=False gives kurtosis = 3 for normal dist
    print(f"{col:35} | {sk:8.2f} | {ku:9.2f}")
```

```python
import math
import seaborn as sns
import matplotlib.pyplot as plt

cols = continuous_cols  # All continuous features
num_plots = len(cols)
plots_per_page = 6
num_pages = math.ceil(num_plots / plots_per_page)

for page in range(num_pages):
    start = page * plots_per_page
    end = min(start + plots_per_page, num_plots)
    page_cols = cols[start:end]
    
    plt.figure(figsize=(15, 8))
    for i, col in enumerate(page_cols, 1):
        plt.subplot(2, 3, i)
        sns.boxplot(y=df1[col], color='skyblue')
        plt.title(f'{col}', fontsize=10)
        plt.tight_layout()
    plt.show()
```

```python
outlier_cols = []

for col in df1.select_dtypes(include='number').columns:
    Q1 = df1[col].quantile(0.25)
    Q3 = df1[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Check if any value is outside bounds
    if ((df1[col] < lower_bound) | (df1[col] > upper_bound)).any():
        outlier_cols.append(col)

print("Columns with outliers:")
print(outlier_cols)
```

```python
df_clipped = df1.copy()

for col in outlier_cols:
    Q1 = df_clipped[col].quantile(0.25)
    Q3 = df_clipped[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_clipped[col] = df_clipped[col].clip(lower=lower_bound, upper=upper_bound)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Copy the clipped DataFrame
df_encoded = df_clipped.copy()

# Define target columns
target_cols = ['tool_condition', 'machining_finalized', 'passed_visual_inspection']

# Encode categorical targets if needed
for col in target_cols:
    if df_encoded[col].dtype == 'object' or str(df_encoded[col].dtype).startswith('category'):
        df_encoded[col] = df_encoded[col].astype('category').cat.codes

# Compute correlation matrix
corr_matrix = df_encoded.corr(numeric_only=True)

# Number of top features to select
N = 10

# Store results
top_features_dict = {}

for target_col in target_cols:
    correlations = corr_matrix[target_col].drop(target_col).abs()
    top_n_features = correlations.sort_values(ascending=False).head(N).index.tolist()
    top_features_dict[target_col] = top_n_features

    print(f"\nTop {N} features most correlated with '{target_col}':")
    print(top_n_features)

# Optional: Visualize correlation heatmap
plt.figure(figsize=(18, 14))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False, fmt=".2f", square=True)
plt.title('Correlation Heatmap (All Continuous Features)', fontsize=16)
plt.show()
```

Preprocessing (Normalization + Sequence Creation)

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Set sequence length
SEQUENCE_LENGTH = 10

# Selected feature columns 
features =  ['M1_CURRENT_FEEDRATE', 'X1_ActualAcceleration', 'X1_ActualPosition', 'X1_ActualVelocity',
 'X1_CommandAcceleration', 'X1_CommandPosition', 'X1_CommandVelocity', 'X1_CurrentFeedback',
 'X1_DCBusVoltage', 'X1_OutputCurrent', 'X1_OutputVoltage', 'Y1_ActualPosition',
 'Y1_CommandPosition', 'Y1_OutputCurrent', 'clamp_pressure', 'feedrate']

# Target columns
targets = ['tool_condition', 'machining_finalized', 'passed_visual_inspection']

# Drop NaNs if any (you may handle them differently if needed)
df_clipped.dropna(subset=features + targets, inplace=True)

# Normalize features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_clipped[features])
scaled_df = pd.DataFrame(scaled_features, columns=features)

# Append targets
for col in targets:
    scaled_df[col] = df_clipped[col].values

# Create sequences

SEQUENCE_LENGTH = 10
num_samples = len(scaled_df) - SEQUENCE_LENGTH

# Preallocate arrays for efficiency
X = np.zeros((num_samples, SEQUENCE_LENGTH, len(features)), dtype=np.float32)
y_tool = np.zeros((num_samples,), dtype=np.int32)
y_machining = np.zeros((num_samples,), dtype=np.int32)
y_visual = np.zeros((num_samples,), dtype=np.int32)

for i in range(num_samples):
    X[i] = scaled_df[features].iloc[i:i+SEQUENCE_LENGTH].values
    y_tool[i] = scaled_df['tool_condition'].iloc[i + SEQUENCE_LENGTH]
    y_machining[i] = scaled_df['machining_finalized'].iloc[i + SEQUENCE_LENGTH]
    y_visual[i] = scaled_df['passed_visual_inspection'].iloc[i + SEQUENCE_LENGTH]


# Train-test split
X_train, X_test, y_tool_train, y_tool_test, y_mach_train, y_mach_test, y_vis_train, y_vis_test = train_test_split(
    X, y_tool, y_machining, y_visual, test_size=0.2, random_state=42, shuffle=False
)
```

 Build LSTM Model

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout

# Input shape
input_layer = Input(shape=(SEQUENCE_LENGTH, len(features)))

# LSTM layers
x = LSTM(128, return_sequences=True)(input_layer)
x = LSTM(64)(x)
x = Dropout(0.2)(x)

# Output layers
tool_output = Dense(3, activation='softmax', name='tool_condition_output')(x)
mach_output = Dense(1, activation='sigmoid', name='machining_finalized_output')(x)
vis_output = Dense(1, activation='sigmoid', name='passed_visual_inspection_output')(x)

# Define model
model = Model(inputs=input_layer, outputs=[tool_output, mach_output, vis_output])
```

Compile the Model

```python
model.compile(
    optimizer='adam',
    loss={
        'tool_condition_output': 'sparse_categorical_crossentropy',
        'machining_finalized_output': 'binary_crossentropy',
        'passed_visual_inspection_output': 'binary_crossentropy'
    },
    metrics={
        'tool_condition_output': 'accuracy',
        'machining_finalized_output': 'accuracy',
        'passed_visual_inspection_output': 'accuracy'
    }
)
```

Train the Model

```python
history = model.fit(
    X_train,
    {
        'tool_condition_output': y_tool_train,
        'machining_finalized_output': y_mach_train,
        'passed_visual_inspection_output': y_vis_train
    },
    epochs=50,
    batch_size=32,
    validation_split=0.1,
    verbose=1
)
```

Evaluate on Test Data

```python
results = model.evaluate(
    X_test,
    {
        'tool_condition_output': y_tool_test,
        'machining_finalized_output': y_mach_test,
        'passed_visual_inspection_output': y_vis_test
    }
)

print("\nTest Evaluation:")
for name, value in zip(model.metrics_names, results):
    print(f"{name}: {value:.4f}")
```
